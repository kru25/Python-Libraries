```python
#Simple Plot
from matplotlib import pyplot as plt
import numpy as np
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
#Multiple plots to add subplot(using add subplot()function)
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(111)  #Main plot
ax1.plot([1,2,3])
ax2 = fig.add_subplot(221, facecolor='y') #facecolor for plot background
ax2.plot([1,2,3])
#Another method to add subplot
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```

```python
#Subplot function
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
fig,ax=plt.subplots(2,2,figsize=(8,10))
ax[0][0].plot(x,2*x)
ax[0][1].plot(x,x**2)
ax[1][0].plot(x,x**3)
ax[1][1].plot(x,x**0.5)
plt.show()
#Using grids in the plots
fig, axes = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25) #Grid paramters(color,ls(line style),lw(Line width))
axes[1].set_title('custom grid')
axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
#setting xticks and yticks
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel("angle")
ax.set_title('sine')
ax.set_xticks([0,2,4,6]) #unit on X
ax.set_xticklabels(['zero','two','four','six']) #labels on X axis
ax.set_yticks([-1,0,1]) #unit on Y,labels on Y axis
plt.show()
```

```python
#Twin axis
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx() #Setting twin axis
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```

```python
#Contour plots=show 3d surface on 2d plane (contour and contourf(Contour fill).countourf,colorbar for colorbar)
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100) #create X
ylist = np.linspace(-3.0, 3.0, 100) #create Y
X, Y = np.meshgrid(xlist, ylist)#creates grid of values
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z) #(x,y,z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.sho
```

```python
#mpltoolkit used for 3D plots
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')#3D projection
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z, 'gray') #plot3D for 3D plots
ax.set_title('3D line plot')
plt.show()
```

```python
#Using log log plot
x = np.linspace(0.01, 5, 10000)
y=x**2
plt.subplot(1,2,1)
plt.plot(x,y,"b-")
plt.text(1,20,"Normal plot") #inserting text on plot (x,y,text)-x,y-coordinates for the text position
plt.annotate('local min', xy=(2, 1), xytext=(3, 1.5),  #text,xy(position),xytext(text position),arrowpops=arrow position
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.subplot(1,2,2)
plt.loglog(x,y,"g-")
```

```python
#Inserting text on plot
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.text(3, 8, 'boxed italics text in data coords', style='italic', #for text in box
bbox = {'facecolor': 'red'})
ax.text(2, 6, r'an equation: $E = mc^2$', fontsize = 15)
ax.text(4, 0.05, 'colored text in axes coords',
verticalalignment = 'bottom', color = 'green', fontsize = 15)
ax.plot([2], [1], 'o')
ax.annotate('annotate', xy = (2, 1), xytext = (3, 4),
arrowprops = dict(facecolor = 'black', shrink = 0.05))
ax.axis([0, 10, 0, 10])
plt.show()
```
