```python
# Transpose of a matrix
a=np.array([[1,2,3],[4,5,6]])
print(a)
print(a.T) #transpose
#iterating row wise
c=a.copy(order="F")
print(c)
for x in np.nditer(c): #iteration column wise
    print(x)
d=a.copy(order="C")
print(d)
for x in np.nditer(d): #iteration row wise
    print(x)
print("Modified array")
for x in np.nditer(d,op_flags=["readwrite"]):
    x[...]=2*x
    print(x)
```

```python
#Traversing each column using flags
print(d)
for x in np.nditer(d,flags=["external_loop"],order="F"):
    print(x)
for x in np.nditer(d,flags=["external_loop"],order="C"): #rowwise
    print(x)
#traversing two arrays together
a = np.arange(0,60,5) 
a = a.reshape(3,4) 

print ('First array is:') 
print (a) 
print ('\n')  

print ('Second array is:')
b = np.array([1, 2, 3, 4], dtype = int) 
print (b) 
print ('\n') 

print ('Modified array is:')
for x,y in np.nditer([a,b]): 
   print("%d:%d" % (x,y))
```

```python
#statistical functions
a=np.array([[1,2,3],[4,5,6]])
print(np.amin(a))
print(np.amax(a))
print(np.amin(a,1))
print(np.amax(a,1))
print(np.amin(a,0)) #Gives min number along axis 0
print(np.amax(a,0))
print(np.ptp(a))
print(np.ptp(a,0)) #gives range along axis 0
print(np.ptp(a,1))
print(np.percentile(a,30))
print(np.percentile(a,30,axis=1))
print(np.percentile(a,30,axis=0))#(array,percentile,axis)
print(np.median(a))
print(np.median(a,1))
print(np.median(a,0))
print(np.mean(a))
print(np.mean(a,0))
print(np.mean(a,1)) #array,axis
print(np.average(a)) # calculates weighed average
print(np.average(a,weights=[[4,5,6],[7,8,9]]))# has argument for weighs after array
print(np.std(a))#standard deviation
print(np.var(a))#variance of the data
```

```python
#sorting of Arrays
a=np.array([[5,4],[3,8]])
print(np.sort(a))
print(np.sort(a,axis=0)) #array,axis,kind,order arguments 
print(np.sort(a,axis=1)) #order is field if the array contains fields
dt = np.dtype([('name', 'S10'),('age', int)]) 
b= np.array([("raju",21),("anil",25),("ravi", 17), ("amar",27)], dtype = dt) 
print(np.sort(b))
print(np.sort(b,axis=0,order="name"))
```

```python
#other sorting functions
x = np.array([3, 1, 2]) #indirect sort provides the indexes of the list in a sorted manner using which array can be
#reconstructed
y=np.argsort(x)
print(x[y]) #reconstructing original array
for i in y:
    print(x[i])
nm = ('raju','anil','ravi','amar') 
dv = ('f.y.', 's.y.', 's.y.', 'f.y.') 
k=np.lexsort((dv,nm)) #assessed using keys
print(k)
for i in k:
    print(nm[i])   
a=np.array([1,3,7,9,0])
print(a[np.argmax(a)]) #index of max element
print(a[np.argmin(a)]) #index of min element
print(np.nonzero(a)) #index of non zero elements
b=np.array([8,5,3,9])
i=np.where(b>8)      
print(b[i]) #to find index of subarray satisfying particular condition
print(np.extract(b>3,b)) #returns elements satisfying condition has 2 arguments-condition and array
```
